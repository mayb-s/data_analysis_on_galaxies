```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#load data
data=pd.read_csv(r"C:\Users\maybe\OneDrive\Desktop\One elective Data Science\cleaned_data.csv")

#check first rows to verify it loaded correctly
print(data.head())
```

```python
#show stats from each column
data.describe()
```

```python
data.info()
```

```python
#types of galaxies, histogram
plt.subplot(2, 2, 1)
plt.hist(data['Type'])
plt.title('Types of Galaxies')
plt.show()
```

```python
#histogram for heliocentric radial velocity
plt.subplot(2, 2, 1)
plt.hist(data['Velocity(km/s)'],bins=20)
plt.title('Velocity Distribution')


#histogram for redshift
#redshift shows a strong correlation with heliocentric radial velocity
#heliocentric radial velocity is speed of galaxy moving away or towards us 
plt.subplot(2, 2, 2)
plt.hist(data['Redshift'], bins=20)
plt.title('Redshift Distribution')
plt.show()
```

```python
#histogram for apparent magnitude
#shows most galaxies have an apparent magnitude of 12 or between 10 and 14
plt.subplot(2, 2, 1)
plt.hist(data['Magnitude'])
plt.title('Apparent Magnitude Distribution')
plt.show()
```

```python
#histoplot for velocity with density line
plt.subplot(2, 1, 1)
sns.histplot(data['Velocity(km/s)'], bins=50, kde=True, element='poly')
plt.title('Velocity Distribution')

plt.show()

#histoplot for redshift with density line
plt.subplot(2, 1, 2)
sns.histplot(data['Redshift'], bins=50, kde=True, element='poly')
plt.title('Redshift Distribution')

plt.show()
```

```python
# Scatter plot of redshift vs apparent magnitude
plt.subplot(2, 1, 1)
sns.scatterplot(x='Redshift', y='Magnitude', data=data)
plt.title('Redshift vs Magnitude')
plt.show()

# Scatter plot of velocity vs apparent magnitude
plt.subplot(2, 1, 2)
sns.scatterplot(x='Velocity(km/s)', y='Magnitude', data=data)
plt.title('Velocity vs Magnitude')
plt.show()

# Scatter plot of velocity vs redshift
plt.subplot(2, 1, 2)
sns.scatterplot(x='Velocity(km/s)', y='Redshift', data=data)
plt.title('Velocity vs Redshift')
plt.show()
```

```python
#types of galaxies vs their velocities
plt.subplot(5, 1, 1)
sns.scatterplot(y='Type', x='Velocity(km/s)', data=data, s=5)
plt.title('Type vs Velocity')
plt.show()

#types of galaxies vs their redshifts
plt.subplot(5, 1, 2)
sns.scatterplot(y='Type', x='Redshift', data=data, s=5)
plt.title('Type vs Redshift')
plt.show()

#types of galaxies vs their magnitudes
plt.subplot(5, 1, 3)
sns.scatterplot(y='Type', x='Magnitude', data=data, s=5)
plt.title('Type vs Magnitude')
plt.show()
```
