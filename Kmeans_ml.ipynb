```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#ml libraries
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
#load data
data=pd.read_csv(r"C:\Users\maybe\OneDrive\Desktop\One elective Data Science\cleaned_data.csv")

#check first rows to verify it loaded correctly
print(data.head())
```

```python
#new critical columns 
critical_columns=['Velocity(km/s)', 'Redshift', 'Magnitude']
data = data[critical_columns]
print(data)
```

```python
#show stats from each column
data.describe()
```

```python
#data is scaled to prevent biased clustering based on range
#scaled value = (value - mean) / standard deviation
#average is changed to zero to center data
scaler=StandardScaler()
scaled_data=scaler.fit_transform(data)
print(scaled_data)
```

```python
# using kmeans, declare number of clusters to fit the data 
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(scaled_data)
```

```python
# define cluster labels and centroids
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

```python
# scatter plot with centroids, magnitude and velocity, columns 0 and 2
plt.scatter(scaled_data[:, 0], scaled_data[:, 2], c=labels, cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 2], s=300, c='red', label='Centroids')
plt.xlabel('Velocity(km/s)')
plt.ylabel('Magnitude')
plt.title('K-means Clustering of Galaxies (Scaled Data)')
plt.legend()
plt.show()
```

```python
# scatter plot with centroids,  magnitude and redshift, columns 1 and 2
plt.scatter(scaled_data[:, 1], scaled_data[:, 2], c=labels, cmap='viridis')
plt.scatter(centroids[:, 1], centroids[:, 2], s=300, c='red', label='Centroids')
plt.xlabel('Redshift')
plt.ylabel('Magnitude')
plt.title('K-means Clustering of Galaxies (Scaled Data)')
plt.legend()
plt.show()
```

```python
# scatter plot with centroids, redshift and velocity, columns 0 and 1
plt.scatter(scaled_data[:, 0], scaled_data[:, 1], c=labels, cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1], s=300, c='red', label='Centroids')
plt.xlabel('Velocity(km/s)')
plt.ylabel('Redshift')
plt.title('K-means Clustering of Galaxies (Scaled Data)')
plt.legend()
plt.show()
```
